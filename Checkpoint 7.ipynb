```python
# Creation of the dataframe
import numpy as np
import pandas as pd
exam_data = {"name" : ["Anastasia", "Dima", "Katherine", "James", "Emily", "Michael", "Matthew", "Laura", "Kevin", "Jonas"], "score" : [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19], "attempts" : [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], "qualify" : ["yes", "no", "yes", "no", "no", "yes", "yes", "no", "no", "yes"]}
labels = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
dataframe1 = pd.DataFrame(exam_data, index = labels)
dataframe1
```

```python
# Question 1
dataframe1.head(3)
```

```python
# Question 2
dataframe1 = dataframe1.dropna()
```

```python
# Question 3
dataframe1[["name", "score"]]
```

```python
# Question 4
k_row = {"name" : "Suresh", "score" : 15.5, "attempts" : 1, "qualify" : "yes"}
dataframe2 = pd.DataFrame(k_row, index = ["k"])
dataframe = pd.concat([dataframe1,dataframe2])
dataframe
```

```python
# Question 5
dataframe.drop("attempts", axis = 1)
```

```python
# Question 6
score_list = dataframe["score"].tolist()
success = []
for score in score_list:
    if score > 10:
        success.append(1)
    else:
        success.append(0)
dataframe['success'] = success
dataframe
```

```python
# Question 7
dataframe.to_csv("my_data.csv")
```
